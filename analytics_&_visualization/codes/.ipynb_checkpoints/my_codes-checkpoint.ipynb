```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Section 1. Import and prepare data

#### 1.1. Import and check ACT data from 2019

```python
# create variable act_19 to store act data from 2019

act_19 = pd.read_csv('../data/act_2019.csv')
print(act_19.info())
print(act_19.head())
```

#### 1.2. Import and check SAT data from 2019

```python
# create variable sat_19 to store sat data from 2019

sat_19 = pd.read_csv('../data/sat_2019.csv')

# Found out that it contains columns EBRW and Math that are not relevant to my research (I only need score and participation).
# So, dropping these columns.
sat_19.drop(columns=['EBRW','Math'], inplace=True)
print(sat_19.info())
print(sat_19.head())
```

#### 1.3. Merge and clean 2019 data. 
The goal is to get one dataframe consisting of: State, ACT score, ACT participation rate, SAT score, SAT participation rate

```python
# Use inner merge to capture data points where both act and sat scores are available and call the new data frame 'tests_precovid'

tests_precovid = pd.merge(act_19,sat_19, how='inner', on=['State'])
print(tests_precovid.info())
print(tests_precovid.head())
```

```python
# Assign descriptive column names

tests_precovid.rename(columns={'Participation':'act_participation', 'Composite':'act_score','Participation Rate':'sat_participation','Total':'sat_score'}, inplace=True)
print(tests_precovid.info())   
```

```python
# Drop percent signs for ACT
c = []
for b in tests_precovid['act_participation']:
    b = int(b.strip('%'))
    c.append(b)
tests_precovid['act_participation'] = c
```

```python
# same for SAT
d = []
for e in tests_precovid['sat_participation']:
    e = int(e.strip('%'))
    d.append(e)
tests_precovid['sat_participation'] = d

print(tests_precovid.head())
```

Done with 2019 data, now work with 2020 data

#### 1.4. Import and check ACT data for 2020

```python
# First, read csv then .info and print head. 
# Found out that the first 3 rows contains titles so going to skip them.
act_20 = pd.read_csv('../data/act_2016_2020.csv', skiprows=3)

### Also found that there are too many information, I only need composite score and participation rate for 2020. So dropping other columns.
act_20.drop(columns=['Unnamed: 13','Composite score','English score','Mathematics score','Reading score','Science score','English score.1','Mathematics score.1', 'Reading score.1','Science score.1','2016',], inplace=True)

### Then, rename the columns so that they are easy to refer to.
act_20 = act_20.rename(columns={'Unnamed: 0':'State','Composite score.1':'act_score', '2020':'act_participation'})

### Found out from .info that there are 3 rows with no score data. So deleting them using iloc method.
act_20[act_20['act_score'].isnull()]
act_20 = act_20.iloc[2:53,:]

### check
print(act_20.head())
print(act_20.info())
```

#### 1.5. Import and check SAT data for 2020

```python
# create variable sat_20 for the data

sat_20 = pd.read_csv('../data/sat_2017_2020_byState.csv')

### Make a list of unwanted columns by eliminating wanted columns from all column names.
a=list(sat_20.columns)
a.remove('State')
a.remove('SAT_Mean_2020')
a.remove('Prcnt_2020')
print(a) # to check, looks correct.

# drop unwanted columns and change the column names to make them descriptive. 
# There is an unwanted row on index 0, going to drop it using iloc.
sat_20.drop(columns=a, inplace=True)
sat_20.rename(columns={'SAT_Mean_2020':'sat_score', 'Prcnt_2020':'sat_participation'}, inplace=True)
sat_20=sat_20.iloc[1:,:]

print(sat_20.info())
print(sat_20.head()) # to check again
# looks good. Continue to EDA
```

#### 1.6. Merge and clean 2020 data

```python
### Merge resulting in errors. Found out that state names on ACT score dataframe has space at the ends of them
### Tried to drop the space using strip() method and replace method (below).

b=list(act_20['State'])
c= []
for s in b:
    str(s).replace(' ','')
    c.append(s)  
print(c) # This code returns a list of states without any change, the whitespace is not removed.
```

```python
### To solve this, I will add a new column of ACT score into the SAT score dataframe. I have checked that the State names are matching.
### Make 2 lists of act score and participation, add them as new columns to sat dataframe, assign a new variable name to it.

d=list(act_20['act_score'].values)
e=list(act_20['act_participation'].values)
sat_20['act_score'] = d
sat_20['act_participation']=e
tests_disrupted = sat_20
print(tests_disrupted.info())
```

Now I have 2 dataframes: tests_precovid and tests_disrupted. Ready to answer research questions.

### Section 2. Answers to Q1: How has the onset of Covid-19 impacted ACT and SAT test scores in general?

#### 2.1. Impacts on ACT 

```python
### Average ACT score in 2019:
def count_mean(x):
    return sum(x)/len(x)
print(count_mean(tests_precovid['act_score'])) # What is average ACT score in 2019

count_mean(tests_disrupted['act_score']) # What is average ACT score in 2020
```

```python
# There is not much change in ACT average between 2019 and 2020 (only 0.06 point)
# Lets plot both scores on histogram

plt.hist(tests_precovid['act_score'], bins=20, alpha=0.5, label= 'precovid')
plt.hist(tests_disrupted['act_score'], bins=20, alpha=0.5, label= 'when covid striked')
plt.legend()
plt.title('ACT scores before and when Covid-19 striked')
plt.xlabel('ACT score')
plt.ylabel('count')
```

**Answer1 to Q1:** Three things are noted in this graph:
1. In general, the graph is flatter in 2020 compared to 2019. 
    This indicates that the number of high and low scorers are more uniform in 2020 compared to 2019. 
2. It can also be noted that, the number of states scoring 19 and 20 (which are in the low range) has dropped about 3 to 4 counts.
    While, the number of states scoring 24 (which is high) has dropped more drastically about 5 counts.
3. Despite disruptions caused by covid onset, score range has increased!
    Highest score was around 25.5 in 2019, but increased to close to 26 in 2020.
    On top of this, the number of states that score higher (between 24.5 and 25) has also increased.

**To answer the question "are there any drops in test score for ACT:** average ACT score did not change much apart from 0.06 point increase in 2020 compared to 2019, 
but highest score rose from 25.5 to 26.

Now, how about SAT?

#### 2.2. Impacts on SAT

```python
# checking average scores
print(count_mean(tests_precovid['sat_score'])) # Average SAT in 2019
count_mean(tests_disrupted['sat_score']) # Average SAT in 2020
```

```python
# Average score has dropped 13 points.
# histogram for SAT scores
plt.hist(tests_precovid['sat_score'], bins=20, alpha=0.5, label= 'precovid')
plt.hist(tests_disrupted['sat_score'], bins=20, alpha=0.5, label= 'when covid striked')
plt.legend()
plt.title('SAT scores before and when Covid-19 striked')
plt.xlabel('SAT score')
plt.ylabel('count')
```

**Answer2 to Q1:** Three points noted on this graph:
1. Most obviously, the number of SAT high scorers (1259 and above) have dropped.
    Highest score was close to 1300 in 2019, but dropped to around 1250 in 2020.
2. There was an increase in the number of students who score poorly (between 950 and 1000) in 2020 compared to 2019.
3. Number of moderate scorers (1100 point) has dropped 3 counts.

**Answer to the question "Are there any drops in score for SAT":** in general, score drops are more observed on SAT than ACT, these include:<br>
     - an obvious drop in highest score scored (from close to 1300 to close to 1250) and<br>
     - a drop of 13 point in average score.


### Section 3. Answer to Q2. It is common knowledge that Covid-19 disruptions had different impacts on different groups of students. Which group of SAT takers -the top five high scorers or the lowest five scorers - has been more impacted when Covid-19 striked and how many percentage drop/rise in scores, compared to the year before, have both groups experienced?

#### 3.1. Who are the top five and bottom five states for SAT before covid?

```python
# Arrange scores from highest to lowest using sort_values

sat_precov1 = tests_precovid.sort_values(['sat_score'], ascending=False)
print(sat_precov1.head(5))

smart_sat_precov = list(sat_precov1['State'][:5])
print(f'SAT smartest five in 2019 are: {smart_sat_precov}') #these are the top 5 states

sat_precov2 = tests_precovid.sort_values(['sat_score'])
low_sat_precov = list(sat_precov2['State'][:5])
print(sat_precov2.head(5))
print(f'SAT lowest 5 scorers in 2019 are: {low_sat_precov}') #and these are the bottom 5
```

#### 3.2. Did the onset of Covid-19 change their positions?

```python
### First, look at who among top ACT scorers and lowest scorers dropped out of the list when covid striked.

sat_disrup1 = tests_disrupted.sort_values(['sat_score'], ascending=False)
smart_sat_disrup = list(sat_disrup1['State'][:5])
print(f'SAT smartest five inspite of Covid-19 onsset are: {smart_sat_disrup}')

sat_disrup2 = tests_disrupted.sort_values(['sat_score'])
low_sat_disrup = list(sat_disrup2['State'][:5])
print(f'SAT lowest five scorers when Covid-19 striked are: {low_sat_disrup}')
      
smart_sat = []
for x in smart_sat_precov:
    if x not in smart_sat_disrup:
        smart_sat.append(x)
print(f'State that dropped out of the top 5 list is: {smart_sat}')

low_sat = []
for i in low_sat_precov:
    if i not in low_sat_disrup:
        low_sat.append(i)
print(f'State that moved up from the bottom 5 list are: {low_sat}')
```

**Observation:** 4 out of 5 top scorers still score highest despite the onset of Covid-19.Only South Dakota dropped out of the list.
States at the bottom positions stayed so when Covid-19 striked.

#### 3.3. How much have scores changed among the top and bottom five when Covid-19 striked? (to answer question: which group is most negatively impacted by the onset of Covid-19?

```python
# Make a new data frame that calculate score difference between 2019 and 2020 for both groups starting with top 5

top_sat_precov = list(sat_precov1['State'][:5])
sat_diff1 = pd.DataFrame(top_sat_precov, columns=['State'])
sat_diff1['sat_precov'] = list(sat_precov1['sat_score'][:5]) # sat_diff1 is the new dataframe

b = list(sat_disrup1[sat_disrup1['State'] == 'South Dakota']['sat_score']) # get the score for south dakota from 2020 data, it is not on 2020 top 5
c = list(sat_disrup1['sat_score'][:5])
c[4] = 24.8
print(c)

sat_diff1['sat_disrup'] = c
sat_diff1['score_difference'] = sat_diff1['sat_disrup'] - sat_diff1['sat_precov'] # add a new col that count score changes from 2020 to 2019
print(sat_diff1)

average_score_difference1 = round(sum(sat_diff1['score_difference']) / 5, 2) # as the variable name suggests
average_score_precov_smart = round(sum(sat_diff1['sat_precov'])/5, 2)
print(average_score_difference1)
d = round(average_score_difference1/average_score_precov_smart, 2)
```

```python
print(f'In average, SAT scores among the top five have dropped by {(average_score_difference1*-1)} point when Covid-19 striked.')
print(f'This drop is equal to {(d*-1)}% difference in average')
```

```python
# 3.4. How much have scores changed among the bottom five when Covid-19 striked
# to answer the question: Among the top five and bottom five, the score of which group is most impacted by the onset of Covid-19?# Make a new data frame that calculate score difference between 2019 and 2020 for both groups
# here i am repeating similar process as above.

e = list(sat_precov2['State'][:5])
sat_diff2 = pd.DataFrame(e, columns=['State'])
sat_diff2['sat_precov'] = list(sat_precov2['sat_score'][:5])

g = list(sat_disrup2['sat_score'][:5])

sat_diff2['sat_disrup'] = g
sat_diff2['score_difference'] = sat_diff2['sat_disrup'] - sat_diff2['sat_precov']
print(sat_diff2)

average_score_difference2 = round(sum(sat_diff2['score_difference']) / 5, 2)
average_score_precov_low = round(sum(sat_diff2['sat_precov'])/5, 2)
print(average_score_difference2)
h = round(average_score_difference2/average_score_precov_low, 3)
```

```python
print(f'In average, SAT scores among the bottom five dropped by {(average_score_difference2*-1)} point when Covid-19 striked.')
print(f'However, this drop is only {(h*-1)}% in average and much smaller compared to the top five group.')
```

**The answer to Q2 is:** The onset of Covid-19 in 2020 has had negative impacts on SAT scores for both groups of states from high achieving and lower achieving margins. This impact is bigger on top five high scorers compared to bottom five group.

### Section 4. Answer to Q3. Which of these geographical groups -the east coast, west coast, or interior states- do better in which tests in 2019? If a state do well in SAT, does it also do well in ACT? (and vice versa) Did the onset of covid-19 in 2020 change this pattern?

```python
### Lets look at our dataframes again: precovid (2019) and disrupted (2020).

print(tests_precovid.head())
print(tests_disrupted.head())
```

```python
# I want to make 2 scatter plots (one for 2019 and one for 2020), each of them shows:
# (i) the relationships between sat and act scores. If a state does well in ACT, does it also do well in SAT and vice versa.
# (ii) which parts of USA do better in which tests?
# To do this, more EDA is needed.

# 4.1. First, I want to divide my 2019 dataframe into 3 dataframes:
# one dataframe representing states on the west of USA, one representing east, 
# and the third one for the states in between west and east.

# Starting with west:
# Make a variable west that holds a list of west states, I want to use this list again later.
west=['Alaska', 'California', ' Hawaii', 'Oregon', 'Washington']
west_coast = tests_precovid[tests_precovid['State'].isin(west)]
print(west_coast.head())
```

```python
# Similar method for east
east=['Connecticut','Delaware','Florida','Georgia','Maine','Maryland','Massachusetts','New Hampshire', 'New Jersey', 'New York', 'North Carolina' 'Rhode Island', 'South Carolina', 'Virginia']
east_coast=tests_precovid[tests_precovid['State'].isin(east)]
print(east_coast.head())                                                        
```

```python
# To get inner states, I use anti-join that subtract west coast and east coast from the whole data.
# Thus, leaving only the inner states.
# Below are the steps (a-e):

step_a = tests_precovid.merge(west_coast, on='State', how='left', indicator=True)
step_b = step_a.loc[step_a['_merge'] == 'left_only', 'State']
step_c = tests_precovid[tests_precovid['State'].isin(step_b)]
```

```python
# continuing from above

step_d = step_c.merge(east_coast, on='State', how='left', indicator=True)
step_e = step_d.loc[step_d['_merge'] == 'left_only', 'State']
inner_states = tests_precovid[tests_precovid['State'].isin(step_e)]
print(inner_states.info())
```

```python
# 4.2. Time to plot precovid tests correlations.

plt.scatter(west_coast['sat_score'], west_coast['act_score'], s=250, c='green', alpha=0.5, edgecolors='white', label='West Coast')
plt.scatter(east_coast['sat_score'], east_coast['act_score'], s=250, c='blue', alpha=0.4, edgecolors='white', label='East Coast')
plt.scatter(inner_states['sat_score'], inner_states['act_score'], s=250, c='magenta', alpha=0.4, edgecolors='white', label='Interior States')
plt.legend(loc='upper right')
plt.xlabel('SAT score')
plt.ylabel('ACT score')
plt.title('Precovid SAT and ACT scores by states')
```

**Answer1 to Q3:**<br>
The plot shows 4 clusters of geographical performance:<br>
Cluster 1: West Coast states doing moderately well on both tests. Between 20 and 23 on ACT and between 1050 and 1150 on SAT.<br>
Cluster 2: East Coast doing low to moderate on SAT (between 950 to 1150) but mostly well on ACT (only 3 states score below 22).<br>
Cluster 3: about half of inner states doing moderately to well on SAT (scoring 1100 and above) but scoring rather low (below 22 points) on ACT.<br>
Cluster 4: the other half of inner states doing rather poorly on SAT(below 1100), but more than half of this group scores above 22 points on ACT.

```python
### Now, lets see how this pattern change/stay the same when covid stiked
```

```python
# 4.3. I want to make a scatter plot that shows relationships between sat and act scores for 2020

# Similar to the above, I am preparing my data into 3 geographical groups.
west=['Alaska', 'California', ' Hawaii', 'Oregon', 'Washington']
west_state = tests_disrupted[tests_disrupted['State'].isin(west)]
print(west_state.head())
```

```python
# east states using similar method as above
east_state = tests_disrupted[tests_disrupted['State'].isin(east)]
print(east_state.head())
```

```python
# inner states using anti-join, same 5 steps as above.

step_f = tests_disrupted.merge(west_state, on='State', how='left', indicator=True)
step_g = step_f.loc[step_f['_merge'] == 'left_only', 'State']
step_h = tests_disrupted[tests_disrupted['State'].isin(step_g)]

step_i = step_h.merge(east_state, on='State', how='left', indicator=True)
step_j = step_i.loc[step_i['_merge'] == 'left_only', 'State']
inner_group = tests_disrupted[tests_disrupted['State'].isin(step_j)]
print(inner_group.info())
```

```python
# 4.4. Time to plot for 2020 data

plt.scatter(west_state['sat_score'], west_state['act_score'], s=250, c='green', alpha=0.5, edgecolors='white', label='West Coast')
plt.scatter(east_state['sat_score'], east_state['act_score'], s=250, c='blue', alpha=0.4, edgecolors='white', label='East Coast')
plt.scatter(inner_group['sat_score'], inner_group['act_score'], s=250, c='magenta', alpha=0.4, edgecolors='white', label='Interior States')
plt.legend(loc='upper right')
plt.xlabel('SAT score')
plt.ylabel('ACT score')
plt.title('SAT and ACT scores by states when Covid-19 striked')
```

**Answer 2 to Q3:** The patterns of clustering look very similar precovid and when covid striked.<br>
There is a similar geographical clustering in 2020 and 2019, despite the disruptions that Covid-19 caused on education.
- West coast states still scored in the ranges of 20-23 for ACT and 1050-1100 for SAT.
- East coast states still scored in the range of 950-1150 on SAT and mostly above 22 on ACT.
- Inner states are still divided into two groups: one did well on SAT but below 22 on ACT and one other group did well on ACT but below 1100 on SAT.

**General answer to q3:** This geographical score divide stays the same in 2020 despite Covid-19 onset.

### Secion 6. Answers to Q5: Which of these geographical groups -the east coast, west coast, or interior states- have higher participation in which tests in 2019? If a state participate highly in ACT, is its participation in SAT lower? (and vice versa) Did the onset of covid change this pattern?

### Section 7. Answers to Q6: Are there any correlations between ACT and SAT scores and participation rates?

```python
# Lets answer this with a heatmap
mask= np.zeros_like(tests_disrupted.corr())
mask[np.triu_indices_from(mask)] = True
tests_disrupted.corr()
plt.figure(figsize=(8,6))
sns.heatmap(tests_disrupted.corr(), annot=True, cmap='coolwarm', mask=mask)
```

**Answer to Q5:** Both heatmap shows high correlations between SAT score and ACT participation and vice versa.
These correlations increase when covid striked.
More research is needed to understand why these parameters are correlated. Maybe only those who scored high submitted their scores?

    - West coast states' participation was moderate to high rate (40%-75%) in SAT and lower rate (20%-50%) in ACT in both 2019 and 2020.
    - Among East coast states, a change in participation rate pattern is noted. However, participation rates are still in the same ranges, which are between 60% and 100% on SAT and mostly below 30% on ACT. 
    - Most obvious change is noted on inner states. Interestingly, in 2019, they had a rather uniform ACT participation rate of 20%, and many of them did not take SAT. In 2020, however, they were divided into 2 groups: one took mostly ACT (60% anc above) and the other took only SAT (100%) and less than 40% ACT.

4. There are high correlations between SAT score and ACT participation and vice versa that increased when Covid-19 striked. However, the correlations between ACT and SAT scores are low. These correlations need further investigation to check whether only those who scored high have submitted their scores.

Building up on these finding, recommendations for further actions include: 
- (i) further research that include other relevant data such as different test policies that different parts of USA have imposed, population sizes, education rates, how badly each states are affected by covid-19; 
- (ii) more research can be done that include data from 2021, knowing that there is a delay response to covid-19 impacts on tests;
- (iii) use these findings to inform future policy decisions that take into considerations geographical characters.
